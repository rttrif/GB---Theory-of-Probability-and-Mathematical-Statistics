## Задача 1.

Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

Среднее значение 

$m = \dfrac{a + b}{2}$

Дисперсия

$D = \dfrac{{(b-a)}^{2}}{12}$

```python
a = 200
b = 800

# Среднее значение 
m = (a + b)/2

# Дисперсия 
D = (b - a)**2 / 12

print(f'Среднее значение: {m}, Дисперсия: {D}')
```

## Задача 4.

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.


Используя формулу $Z = \left(\frac{X - \mu}{\sigma}\right)$ и таблицу накопленного нормального распределения получаем:

```python
def z(x):
    Mx = 174
    sigma = 8
    return (x - Mx) / sigma

print('Z(150) =', z(150))
print('Z(158) =', z(158))
print('Z(166) =', z(166))
print('Z(182) =', z(182))
print('Z(190) =', z(190))
print('Z(198) =', z(198))
```

```python
# Z(182) = 1.0
print('а). больше 182 см:', 1 - 0.8413)

# Z(190) = 2.0
print('б). больше 190 см:', 1 - 0.9772)

# Z(166) = -1.0
# Z(190) = 2.0
print('в). от 166 см до 190 см:', 0.9772 - 0.1587)

# Z(166) = -1.0
# Z(182) = 1.0
print('г). от 166 см до 182 см:', 0.8413 - 0.1587)

# Z(158) = -2.0
# Z(190) = 2.0
print('д). от 158 см до 190 см:', 0.9772 - 0.0228)

# Z(150) = -3.0
# Z(190) = 2.0
print('е). не выше 150 см или не ниже 190 см:', 0.0014 + 1 - 0.9772)

# Z(150) = -3.0
# Z(198) = 3.0
print('ё). не выше 150 см или не ниже 198 см:', 0.0014 + 1 - 0.9986)

# Z(166) = -1.0
print('ж). ниже 166 см.:', 0.1587)
```
